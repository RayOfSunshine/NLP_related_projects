```python
import numpy as np
import struct
import os
```

```python
def load_pretrained_embedding(path):
    """
    读取token_vec_300.bin中训练好的词向量
    返回为字典格式：，key为字，value为其300维的向量（array类型，每个数字为float32），
    """
    embeddings_dict = {}
    with open(path, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.strip().split(' ')
            if len(values) < 300:
                continue
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_dict[word] = coefs
    print('Found %s word vectors.' % len(embeddings_dict))
    return embeddings_dict
```

```python
def build_embedding_matrix(path):
    embedding_dict = load_pretrained_embedding(path)
    # 假设存在 word_dict
    word_dict = {'字': 0, '成': 1, '压': 2}
    vocab_size = len(word_dict)
    embedding_matrix = np.zeros((vocab_size + 1, 300))  # 预先生成（词表长度+1，词向量维度）的 embedding_matrix
    for word, i in word_dict.items():
        embedding_vector = embedding_dict.get(word)
        if embedding_vector is not None:
            # 如果词表中的词在embedding_dict中是有出现的，则将这个词的对应的词向量添加进embedding_matrix；不出现则表示为0
            embedding_matrix[i] = embedding_vector
    return embedding_matrix
```

```python
path = './token_vec_300.bin'
emb_mat = build_embedding_matrix(path)
print(emb_mat.shape)
print(emb_mat)
```
